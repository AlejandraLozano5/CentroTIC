```python
import numpy
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

Para convertir los datos del sensor en ppm se realizan los siguientes pasos:

* Los datos sensor de gas corresponden al voltaje medido realmente corresponde al divisor de tensión

$V_o = V_{ref}\left(\frac{R_l}{Rl+R_s}\right)$      $R_l$: es la resistencia de carga generalmente 1k


* se despeja Rs


$R_s = \frac{R_l(V_{ref}-V_o)}{V_o}$

* Con Rs encontrada ya podemos calcular el ppm

$ppm = \left(\frac{AR_o}{R_s}\right)^\frac{1}{b}$   

$A$ y $b$ son constantes que se sacan de la curva $y = Ax^{-b}$

$Ro$: es una constante que se calcula a 1000 ppm del gas, o en el punto de saturación del gas

# Sensor MQ4 para medir CH4

```python
def func_ecn(x,a,b):
    return a*x**(-b)

ppm = numpy.array([200, 300, 400, 500, 900, 1000,2000, 3000, 4000,6000,7000,10000]) #x
Rs_Ro = numpy.array([1.85,1.55,1.33, 1.2, 1.08, 1, 0.79, 0.61, 0.6,0.5,0.49, 0.45]) #y

popt, pcov = curve_fit(func_ecn, ppm, Rs_Ro)
ecn = r"$y= {0:.2f}x$".format(popt[0])+"^{0:.2f}".format(-popt[1])
```

```python
plt.figure(figsize=(10,5))
plt.plot(ppm, func_ecn(ppm, *popt), 'g--', label=ecn)
plt.plot(ppm,Rs_Ro,'o',label="datos curva CH4")
plt.xlabel("ppm")
plt.ylabel("Rs/Ro")
plt.title("Curva del sensor MQ4 para CH4")
plt.legend()
plt.grid(True)
```

# Sensor ULPSM-SO2 para medir SO2

Es un sensor lineal que de acuerdo al datasheet https://www.spec-sensors.com/wp-content/uploads/2016/10/ULPSM-SO2-968-006.pdf tiene la siguiente fórmula:

$C_x=\frac{1}{M}(V_{gas}-V_{gas0})$

donde $V_gas$ es la lectura del sensor y $V_gas0$ es la lectura del sensor cuando no hay gas

$M (V/ppm) = 4.94\times100\times10^{-9}\times10^3$

$M=0.000494$

# Sensor MiCS-2714 para medir NO2
